# Functions


```python
def calculate_age_to_age_factors(cumulative_triangle):
    """
    Calculate individual age-to-age factors for every accident year.

    Parameters
    ----------
    cumulative_triangle : pd.DataFrame
        Cumulative loss triangle

    Returns
    -------
    pd.DataFrame
        Age-to-age factor matrix
    """

    factors = pd.DataFrame(index=cumulative_triangle.index)

    for col in range(cumulative_triangle.shape[1] - 1):

        current = cumulative_triangle.iloc[:, col]
        nxt = cumulative_triangle.iloc[:, col + 1]

        factors[f"{col+1}->{col+2}"] = nxt / current

    return factors
```

```python
def calculate_ldfs(cumulative_triangle):
    """
    Calculate volume-weighted Chain Ladder LDFs.

    Parameters
    ----------
    cumulative_triangle : pd.DataFrame

    Returns
    -------
    pd.DataFrame
    """

    results = []

    for col in range(cumulative_triangle.shape[1] - 1):

        current = cumulative_triangle.iloc[:, col]
        nxt = cumulative_triangle.iloc[:, col + 1]

        valid = current.notna() & nxt.notna()

        ldf = nxt[valid].sum() / current[valid].sum()

        results.append({
            "Development": f"{col+1} -> {col+2}",
            "LDF": ldf
        })

    return pd.DataFrame(results)
```

```python
def calculate_cdfs(ldf_df):
    """
    Calculate cumulative development factors.
    """

    ldfs = ldf_df["LDF"].values

    cdfs = []

    running_product = 1

    for ldf in reversed(ldfs):
        running_product *= ldf
        cdfs.append(running_product)

    cdfs.reverse()

    result = ldf_df.copy()
    result["CDF"] = cdfs

    return result
```

```python
def get_latest_diagonal(cumulative_triangle):

    results = []

    for ay in cumulative_triangle.index:

        row = cumulative_triangle.loc[ay]

        latest_col = row.last_valid_index()

        latest_value = row[latest_col]

        results.append({
            "AccidentYear": ay,
            "DevelopmentAge": int(latest_col),
            "LatestValue": latest_value
        })

    return pd.DataFrame(results)
```

```python
def estimate_ultimate_losses(latest_diag, cdf_lookup):

    results = []

    for _, row in latest_diag.iterrows():

        age = int(row["DevelopmentAge"])

        latest = float(row["LatestValue"])

        cdf = cdf_lookup.get(age)

        if cdf is None:
            raise ValueError(f"No CDF found for development age {age}")

        ultimate = latest * cdf
        reserve = ultimate - latest

        results.append({
            "AccidentYear": row["AccidentYear"],
            "DevelopmentAge": age,
            "LatestValue": latest,
            "CDF": cdf,
            "Ultimate": ultimate,
            "Reserve": reserve
        })

    return pd.DataFrame(results)
```

```python
def build_cdf_lookup(cdf_table):

    lookup = {
        i: cdf
        for i, cdf in enumerate(
            cdf_table["CDF"],
            start=1
        )
    }

    lookup[len(cdf_table) + 1] = 1.0

    return lookup
```

# Paid triangle

```python
paid_link_ratios = calculate_age_to_age_factors(paid_cum)

paid_link_ratios.head()
```

```python
paid_selected_ldfs = calculate_ldfs(paid_cum)
paid_selected_ldfs
```

```python
paid_cdf_table = calculate_cdfs(paid_selected_ldfs)

paid_cdf_table
```

```python
paid_cdf_lookup = {}

for i, cdf in enumerate(
    paid_cdf_table["CDF"],
    start=1
):
    paid_cdf_lookup[i] = cdf

paid_cdf_lookup[len(paid_cdf_table) + 1] = 1.0
```

```python
paid_latest_diag = get_latest_diagonal(paid_cum)

paid_latest_diag.head()
```

```python
ultimate_table_paid = estimate_ultimate_losses(
    paid_latest_diag,
    paid_cdf_lookup
)
ultimate_table_paid["ReservePct"] = (
    ultimate_table_paid["Reserve"] /
    ultimate_table_paid["Ultimate"]
)
ultimate_table_paid
```

# Incurred Claims Triangle

```python
incurred_link_ratios = calculate_age_to_age_factors(incurred_cum)
incurred_link_ratios.head()
```

```python
incurred_selected_ldfs = calculate_ldfs(incurred_cum)
incurred_selected_ldfs
```

```python
incurred_cdf_table = calculate_cdfs(incurred_selected_ldfs)

incurred_cdf_table
```

```python
incurred_cdf_lookup = {}

for i, cdf in enumerate(
    incurred_cdf_table["CDF"],
    start=1
):
    incurred_cdf_lookup[i] = cdf

incurred_cdf_lookup[len(incurred_cdf_table) + 1] = 1.0
```

```python
incurred_latest_diag = get_latest_diagonal(incurred_cum)
incurred_latest_diag.head()
```

```python
ultimate_table_incurred = estimate_ultimate_losses(
    incurred_latest_diag,
    incurred_cdf_lookup
)
ultimate_table_incurred["ReservePct"] = (
    ultimate_table_incurred["Reserve"] /
    ultimate_table_incurred["Ultimate"]
)
ultimate_table_incurred
```

```python
comparison = pd.DataFrame({
    "PaidUltimate":
        ultimate_table_paid["Ultimate"],

    "IncurredUltimate":
        ultimate_table_incurred["Ultimate"]
})

comparison["Difference"] = (
    comparison["PaidUltimate"]
    - comparison["IncurredUltimate"]
)

comparison
```

```python
# ==================================================
# RESERVE DIAGNOSTICS AGENT
# ==================================================
```

```python
diag = ultimate_table_paid.merge(
    ultimate_table_incurred,
    on="AccidentYear",
    suffixes=("_Paid", "_Incurred")
)
```

```python
diag["Difference"] = (
    diag["Ultimate_Paid"]
    - diag["Ultimate_Incurred"]
)

diag["Pct_Difference"] = (
    diag["Difference"]
    / diag["Ultimate_Incurred"]
)
```

```python
def divergence_flag(x):

    x = abs(x)

    if x > 0.25:
        return "FAIL"

    elif x > 0.10:
        return "WARNING"

    return "PASS"
```

```python
diag["DivergenceStatus"] = (
    diag["Pct_Difference"]
    .apply(divergence_flag)
)
```

```python
diag[
    [
        "AccidentYear",
        "Ultimate_Paid",
        "Ultimate_Incurred",
        "Difference",
        "Pct_Difference",
        "DivergenceStatus"
    ]
]
```

```python
diag.loc[
    diag["DivergenceStatus"] != "PASS",
    [
        "AccidentYear",
        "Pct_Difference",
        "DivergenceStatus"
    ]
]
```

```python
diag = diag.merge(
    paid_latest_diag[
        ["AccidentYear", "DevelopmentAge"]
    ],
    on="AccidentYear",
    how="left"
)
```

```python
diag[
    ["AccidentYear", "DevelopmentAge"]
]
```

```python
diag["CDF"] = (
    diag["DevelopmentAge"]
    .map(paid_cdf_lookup)
)
```

```python
diag[
    ["AccidentYear", "DevelopmentAge", "CDF"]
]
```

```python
diag["RemainingDevelopment"] = (
    diag["CDF"] - 1
)
```

```python
def immature_flag(x):

    if x > 0.50:
        return "IMMATURE"

    return "MATURE"
```

```python
diag["ImmaturityStatus"] = (
    diag["RemainingDevelopment"]
    .apply(immature_flag)
)
```
